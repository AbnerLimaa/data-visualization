```python
from matplotlib import pyplot

movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side History"]
num_oscars = [5, 11, 3, 8, 10]
x_coordinates = [i for i, _ in enumerate(movies)]

pyplot.bar(x_coordinates, num_oscars)
pyplot.ylabel("Number of academy awards")
pyplot.title("Movies")
pyplot.xticks([i for i, _ in enumerate(movies)], movies)
pyplot.show()
```

```python
from collections import Counter

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
decile = lambda grade: grade // 10
histogram = Counter(decile(grade) for grade in grades)

pyplot.bar([key * 10 for key in histogram.keys()], histogram.values(), 8)
pyplot.axis([-5, 105, 0, 5])
pyplot.xticks([10 * i for i in range(11)])
pyplot.xlabel("Decile")
pyplot.ylabel("Number of students")
pyplot.title("Distribution of Exam 1 Grades")
pyplot.show()
```

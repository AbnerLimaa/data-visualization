```python
from matplotlib import pyplot

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

pyplot.plot(years, gdp, color='green', marker='o', linestyle='solid')
pyplot.title("Nominal GDP")
pyplot.ylabel("Billions of $")
pyplot.show()
```

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
x_coordinates = [i for i, _ in enumerate(variance)]

pyplot.plot(x_coordinates, variance, 'g-', label="variance")
pyplot.plot(x_coordinates, bias_squared, 'r-.', label="bias^2")
pyplot.plot(x_coordinates, total_error, 'b:', label="total error")
pyplot.legend(loc=9)
pyplot.xlabel("model complexity")
pyplot.title("The Bias Variance Tradeoff")
pyplot.show()
```
